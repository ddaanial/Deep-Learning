# Single-Cell RNA Sequencing

Single-cell RNA sequencing (scRNA-seq) is a powerful technique used in molecular biology to study gene expression at the single-cell level. It allows researchers to profile the transcriptome of individual cells, providing insights into cellular heterogeneity and identifying distinct cell types within a tissue or sample.

By analyzing gene expression at the single-cell level, scRNA-seq enables researchers to gain insights into the diversity of cell types, identify rare cell populations, characterize cell states, and study cellular dynamics in various biological processes, such as development, disease progression, or response to treatments.

It's important to note that scRNA-seq is a rapidly evolving field, and new methods and technologies are continuously being developed to improve sensitivity, throughput, and the ability to capture additional molecular features beyond gene expression, such as chromatin accessibility or protein levels.

*   Compare the typical bulk RNA-seq vs scRNA-seq and state what advantages scRNA-seq data could provide to analyse.

1. Sample Complexity:
   - Bulk RNA-seq: Measures gene expression from a population of cells, resulting in an averaged profile.
   - scRNA-seq: Provides gene expression data at the single-cell level, allowing analysis of individual cellular heterogeneity within a population.

2. Cell Resolution:
   - Bulk RNA-seq: Provides average expression levels across all cells in a sample, which can mask cell-to-cell variations.
   - scRNA-seq: Allows identification of distinct cell types, rare cell populations, and cellular states, enabling detailed characterization of cell subpopulations.

3. Gene Expression Variability:
   - Bulk RNA-seq: Cannot distinguish whether observed gene expression changes occur uniformly across cells or are driven by specific cell subsets.
   - scRNA-seq: Captures cell-to-cell variability, facilitating the identification of cell-specific gene expression patterns and understanding of cellular dynamics.

4. Discovery of Novel Transcripts and Splicing Variants:
   - Bulk RNA-seq: Provides an overview of common transcripts and isoforms expressed in a sample.
   - scRNA-seq: Enables the detection of rare or lowly expressed transcripts, alternative splicing events, and cell-type-specific isoforms.

5. Pathway Analysis and Regulatory Networks:
   - Bulk RNA-seq: Suitable for identifying differentially expressed genes between conditions and performing pathway analysis on a population scale.
   - scRNA-seq: Allows reconstruction of cellular regulatory networks, inference of cell-cell interactions, and identification of rare cell types involved in specific pathways.

6. Trajectory and Developmental Analysis:
   - Bulk RNA-seq: Provides a snapshot of gene expression dynamics but lacks information about cell lineage relationships.
   - scRNA-seq: Enables the inference of cellular trajectories, identification of transitional states during development, and characterization of lineage-specific gene expression changes.

7. Spatial Transcriptomics:
   - Bulk RNA-seq: Provides ensemble-level gene expression information without spatial context.
   - scRNA-seq: Can be integrated with spatial transcriptomics techniques to map gene expression patterns within intact tissues and infer spatial organization.

*   What are the key steps involved in performing a scRNA-seq experiment?



1. Experimental Design: Define the primary objectives of the study and consider factors such as cell types, developmental stages, or conditions to be compared. Determine the number of cells to be sequenced based on statistical power calculations.

2. Sample Preparation: Isolate and prepare single cells for sequencing using appropriate protocols. This may involve tissue dissociation, cell sorting, or single-cell capture techniques. Maintain cell viability and integrity throughout the process.

3. Library Preparation: Generate sequencing libraries from individual cells. This typically includes reverse transcription of RNA into complementary DNA (cDNA), amplification, and addition of sample-specific barcodes to enable multiplexing. Consider using unique molecular identifiers (UMIs) to mitigate amplification biases.

4. Quality Control: Assess library quality and quantity using methods like fluorometric quantification, qPCR, or bioanalyzer analysis. Validate the libraries for proper representation of target cell populations.

5. Sequencing: Submit the prepared libraries for high-throughput sequencing using platforms such as Illumina, 10x Genomics, or Drop-seq. Aim for sufficient sequencing depth to capture rare transcripts adequately.

6. Data Preprocessing: Perform initial data preprocessing steps, including read alignment to a reference genome, removal of low-quality reads, and PCR duplicate removal. Use specialized software packages designed for scRNA-seq data, such as Cell Ranger, Seurat, or Scanpy.

7. Quality Control of Raw Data: Assess the quality of the raw sequencing data by examining metrics such as read distribution, gene body coverage, and mitochondrial gene content. Exclude poor-quality cells with high levels of contamination or technical artifacts.

8. Normalization: Apply normalization methods to account for cell-to-cell variation in sequencing depth and other technical effects. Common normalization methods include total count scaling, library size normalization, or spike-in controls.

9. Cell Filtering: Filter out low-quality cells, doublets, or empty droplets based on various quality metrics such as the number of detected genes or unique molecular identifiers. Consider excluding cells with high mitochondrial gene content as they may indicate cell stress or contamination.

10. Dimensionality Reduction: Reduce the high-dimensional scRNA-seq data to a lower-dimensional representation using techniques such as principal component analysis (PCA) or t-distributed stochastic neighbor embedding (t-SNE). This step facilitates downstream analysis and visualization.

11. Clustering: Perform unsupervised clustering to identify distinct cell populations based on gene expression patterns. Use clustering algorithms like K-means, hierarchical clustering, or graph-based approaches. Validate the clusters using known marker genes or external references.

12. Differential Expression Analysis: Identify differentially expressed genes between cell populations to understand their functional differences. Statistical methods like the negative binomial test (e.g., implemented in packages like DESeq2 or edgeR) are commonly used for this analysis.

13. Trajectory Analysis: If the experimental design involves developmental or temporal processes, reconstruct cellular trajectories using inference algorithms like Monocle, Slingshot, or PAGA. This allows the characterization of cellular states and lineage relationships.

14. Functional Analysis: Perform gene ontology enrichment analysis, pathway analysis, or other functional annotations to gain insights into the biological processes and pathways associated with specific cell populations or states.

15. Data Integration: Optionally, integrate scRNA-seq data from multiple experiments or platforms to increase sample size or compare datasets. Several integration methods, such as Harmony, Seurat v3, or Scanorama, have been developed for this purpose.

16. Visualization: Visualize the scRNA-seq data using plots like heatmaps, violin plots, or feature plots to explore gene expression patterns across cell populations or states. Interactive visualization tools like R Shiny or CellBrowser can aid in data exploration.

*   Explain the difficulties with scRNA-seq data acquisition. Specifically, explain what are doublets and dropouts.


Doublets: Doublets occur when two or more cells are inadvertently captured and processed together as a single cell during the experimental process. This can happen due to technical issues such as cell clumping or low cell capture efficiency. The resulting scRNA-seq data from doublets can be misleading since gene expression profiles from different cells are combined. Doublets can lead to incorrect identification of cell types, inaccurate assessment of gene expression patterns, and distorted downstream analyses. Various computational methods have been developed to detect and remove doublets from scRNA-seq datasets, but completely eliminating them remains challenging.

Dropouts: Dropouts refer to the phenomenon in scRNA-seq where the expression of certain genes is not detected in individual cells, even though the gene might be expressed in reality. This occurs due to technical limitations associated with the scRNA-seq workflow, including low RNA capture efficiency and amplification biases. Dropouts can lead to false negatives, where true gene expression activity is missed, resulting in an incomplete and sparse representation of the transcriptome. Consequently, genes that are expressed at low levels or in a subset of cells may be underestimated or completely undetected. Dealing with dropouts requires careful consideration during data analysis, including imputation methods to estimate missing values and statistical approaches to handle the inherent noise in the data.

  No, single-cell RNA sequencing (scRNA-seq) specifically focuses on capturing and profiling gene expression at the single-cell level. It provides information about the transcriptome of individual cells, allowing researchers to identify and characterize different cell types, understand cellular heterogeneity, and study gene regulation.

# Pre-processing

Preprocessing plays a crucial role in single-cell RNA sequencing (scRNA-seq) data analysis. It involves a series of computational steps aimed at transforming raw sequencing data into a format suitable for downstream analysis. Preprocessing tasks include quality control, filtering, normalization, gene expression estimation, and handling technical artifacts. By carefully preprocessing the data, researchers can improve the accuracy, reliability, and interpretability of scRNA-seq results, enabling the identification of meaningful biological patterns, cell types, and regulatory networks.

## Normalization

Up to this point the data is only available as a count matrix. Counts are representative of molecules that were captured in the scRNA-seq experiment. As not all mRNA molecules in a cell are captured, there is a variability in the total number of counts detected between cells that results from both the number of molecules that were in the cells, and the sampling. As we cannot assume that all cells contain an equal number of molecules (cell sizes can differ substantially), we have to estimate the number of molecules that were initially in the cells. In fact, we don't estimate the exact number of molecules, but instead estimate cell-specific factors that should be proportional to the true number of molecules. These are called size factors. Normalized expression values are calculated by dividing the measured counts by the size factor for the cell. [1]

Based on comparisons of normalization methods for differential testing (personal communication from Beathe Vieth), and batch correction [Buettner et al., 2019], the normalization method implemented in the scran package performs best. This method requires a coarse clustering input to improve size factor esimation performance. Thus, we need to use a simple preprocessing approach and cluster the data at a low resolution to get an input for the size factor estimation. The basic preprocessing includes assuming all size factors are equal (library size normalization to counts per million - CPM) and log-transforming the count data. [1] *

*You may use this approach (which is of higher quality), or you can just simply use scanpy and normalize assuming all size factors are equal. e.g. ```sc.pp.normalize_total(adata, target_sum=1e6)```


Before normalizing the data, always ensure that a copy of the raw count data is kept in a separate AnnData object. This allows us to use methods downstream that require this data as input. [1]

The normalized count data must be log-transformed with an offset of 1. This is performed to normalize the data distributions. The offset of 1 ensures that zero counts map to zeros. [1]

It is good practice to store a copy of the current state of your anndata in `.raw` part of your anndata object as log-normalised data for statistical testing: `adata.raw = adata`

[1] `Luecken MD, Theis FJ. Current best practices in single-cell RNA-seq analysis: a tutorial. 2019`

*   Can you tell us about the problems that could occur, if we do not Normalize, or log-transform our data?


1. Differential sequencing depth: Each cell in scRNA-seq experiments can have varying sequencing depths due to technical factors. Without normalization, highly expressed genes can dominate the data, masking the expression levels of lower-abundance genes. This makes it difficult to compare gene expression across cells accurately.

2. Batch effects: scRNA-seq experiments often involve multiple batches or experimental conditions. If normalization is not performed, these batch effects can confound the biological signal, leading to inaccurate interpretations. Normalization is crucial for removing batch effects to ensure meaningful downstream analyses.

3. Cell variability: Cells within a sample can have inherent biological variability, and without normalization, this can be further magnified. To compare gene expression between cells accurately, normalization helps to remove unwanted variations originating from technical factors, allowing for reliable comparisons.

4. Log-transformation for variance stabilization: scRNA-seq data often follows a log-normal distribution, where the majority of genes' expression levels are low, while a few genes may have extremely high expression. By log-transforming the data, the variance between low- and high-expressed genes can be better balanced, enabling appropriate downstream statistical analyses.

*   Please explain one of your chosen algorithms in detail.

The ComBat algorithm is a method used to harmonize and remove batch effects from high-dimensional datasets. Batch effects refer to variations in data due to technical aspects like different lab conditions, machines, or date of data collection, rather than the biological signal of interest. These batch effects can lead to misleading results and make it challenging to compare and integrate data from different sources.

The primary goal of ComBat is to adjust the data in such a way that the biological signal is preserved, while the non-biological variation attributed to batch effects is minimized. ComBat achieves this by assuming that the data can be decomposed into two components: the biological signal and the batch effects.

Here's how the ComBat algorithm works step by step:

1. Input: The algorithm takes as input a high-dimensional dataset, typically represented as a matrix with samples as rows and features (genes, proteins, etc.) as columns. Additionally, it requires information about the batch or experimental condition of each sample.

2. Model fitting: ComBat fits a linear model to each feature using the batch information as a covariate. It estimates the effect of the batch on each feature while accounting for the biological signal.

3. Empirical Bayes estimation: ComBat uses an empirical Bayes approach to estimate the batch effect parameters. It borrows information across features to estimate a common mean and variance for batch effects, reducing noise and improving estimation accuracy.

4. Harmonization: Next, ComBat applies a batch adjustment procedure to harmonize the dataset. It subtracts the estimated batch effects from each feature across samples, effectively removing the batch-specific differences. The biological signal remains intact, while the unwanted variation is eliminated.

5. Model prediction: After harmonization, ComBat regenerates the dataset for each feature. It uses the fitted linear models to predict the expected values of features in the absence of batch effects.

6. Output: The final output of the algorithm is a harmonized dataset with reduced batch effects.

*   Please mention some of the problems that could occur in further analysis and work, if batch correction is not performed

1. Batch effects: Batch effects refer to systematic technical variations that occur during data acquisition or processing, leading to differences between batches of samples. Without batch correction, these effects can confound the interpretation of the data by artificially inflating or reducing the observed variability between samples.

2. Inconsistent results: Without batch correction, the results obtained from different batches may not be directly comparable. This could hinder the ability to draw meaningful conclusions or generalize findings across different batches.

3. Misinterpretation of biological variation: Batch effects can overshadow or mask the true biological variation present in the data. Without correcting for these effects, it can be difficult to differentiate between batch-specific changes and actual biological differences.

4. Reduced statistical power: Batch effects can introduce noise into the data, reducing the statistical power to detect true associations or effects. This can lead to underpowered studies and potentially miss important discoveries.

5. False positives/negatives: Failure to perform batch correction may result in the identification of false positives or false negatives, leading to erroneous conclusions or missed important findings.

6. Reproducibility issues: In the absence of batch correction, the reproducibility of the analysis can be compromised. Other researchers trying to replicate or validate the results may face difficulties due to the unaccounted batch effects.

*   Please explain, how the algorithm you used, is choosing the Highly Variable Genes.

Based on the following steps:

1. Normalize the gene expression values using the Seurat package's NormalizeData() function. This step scales the expression values to a comparable range and uses the natural log transformation.

2. Estimate the mean and variance of the log-transformed expression values for each gene in each cell. This is done using the FindVariableFeatures() function in Seurat.

3. Fit a curve to the relationship between the mean and variance of log expression for each gene. This step is performed using a robust linear regression model.

4. Calculate a gene-specific dispersion statistic, which measures the level of biological variability relative to technical variability. A high dispersion value indicates higher biological variability, making the gene a potential candidate for being highly variable.

5. Determine highly variable genes based on the dispersion statistic, mean expression, and a minimum detection threshold. Genes with high dispersion and expressed in a sufficient number of cells are considered highly variable.

*   Why is using HVGs important while doing inference and learning on the data? Why not use the whole set of genes?

1. Noise reduction: The whole set of genes in a dataset may include genes that do not vary much across different samples or conditions. Including these genes in the analysis can increase the noise in the dataset, making it harder to identify patterns or detect meaningful differences. By focusing on highly variable genes, which exhibit more significant changes between samples or conditions, the noise in the data can be reduced.

2. Computational efficiency: Most machine learning and inference algorithms operate on numerical representations of data, such as matrices. Including the entire set of genes can lead to very high-dimensional datasets, which can pose computational challenges and increase the complexity of analysis. By selecting highly variable genes, the dimensionality of the dataset is reduced, making computations more tractable and efficient.

3. Biologically relevant information: Highly variable genes often correspond to genes that are involved in important biological processes or exhibit distinct expression patterns. These genes may represent key regulators, biomarkers, or drivers of specific biological phenomena. By focusing on these genes, researchers can gain insights into the underlying mechanisms driving the observed patterns or differences in the data.


# Clustering

Clustering is a key step in scRNA-seq analysis since it can give us insights to better understant our data, find relations between different conditions in cells, identify the cell types and the state of the cell.

Among different algorithms available to perform clustering, `Louvain` algorithm is a common algorithm in single cell analysis. (The Louvain algorithm has been proposed for single-cell analysis by [[Levine15]](https://scanpy.readthedocs.io/en/stable/references.html#levine15).)

In `scanpy`, louvain is performed on the highly variable gene data, dimensionality reduced by PCA, and embedded into a KNN graph by default. So make sure to run `pca`, `neighbors`, and `umap (to visualize)` after computing the HVGs and before running `louvain`. (see `sc.pp.pca()` and `sc.pp.neighbors()`)

You can try different `resolutions` for `louvain` to find a fit clustering. (Visualize each clustering regarding to each resolution)

````python
resolutions = [0.2, 0.5, 1.0] 

for resolution in resolutions:
    sc.tl.louvain(adata_combat, resolution=resolution)
    if 'my_clustering' not in adata_combat.obs.keys():
        adata_combat.obs['my_clustering'] = adata_combat.obs['louvain']
    else:
        adata_combat.obs['my_clustering'] = adata_combat.obs['my_clustering'].astype(str)

    sc.pl.umap(adata_combat, color=['louvain'], legend_loc='on data', title=f'{resolution}-resolution')


````

## Marker Genes Identification

To gain informative data from our clustering, we can identify the marker genes in each cluster. We expect each cluster to be able to contain a group of specific cells (e.g. based on cell types, drugs, other perturbations).

Therefore, we can also expect to find a number of genes in each cluster that are up-regulated compared to all other clusters. This is performed by computing a ranking for the highly differential genes in each cluster and comparing them to each other.

This task can be done using the `sc.tl.rank_genes_groups` function of `scanpy` which uses `t-test` by default and is performed on the `.raw` dataset which is uncorrected and contains all genes. All genes are taken into account, as any gene may be an informative marker.

You need to specify your clustering's name from `obsm` so scanpy can rank the marker genes with regarding to those clusters (`groupby`).

You can then plot the result of the marker gene analysis using `sc.pl.rank_genes_groups`

````python
sc.tl.rank_genes_groups(adata_combat, groupby='my_clustering')
sc.pl.rank_genes_groups(adata_combat, groupby='my_clustering')
````
